# Merge basics with Movie Ratings tables based on movie_id
- Start with movie_ratings table
- Pull in movie name using a merge (left join)

```python
movies_with_ratings = pd.merge(movie_ratings, movie_basics, left_on='movie_id', right_on='movie_id')
movies_with_ratings
```

### Analyze the movies where we have ratings
- When were the movies from?

```python
# import matplotlib.pyplot as plt
# fig, ax = plt.subplots()
# ax.hist(df['start_year'], bins=10);

pd.pivot_table(movies_with_ratings, values="primary_title", aggfunc="count", index="start_year")
```

```python
movies_with_ratings.describe()
```

```python
movies_with_ratings['numvotes']
```

### Focus on movies with a sample of votes (ratings):  
- The more number of votes a movie has, the better we can feel about the average rating.  Votes totals of 200 or higher

```python
vote_threshold = movies_with_ratings['numvotes'].mean()
focusMovies = movies_with_ratings[(movies_with_ratings['numvotes'] >= vote_threshold)]
focusMovies.sort_values('averagerating', ascending=False)
```
