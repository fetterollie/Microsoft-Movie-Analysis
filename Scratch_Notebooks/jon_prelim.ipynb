## Trying to merge

The data frames I am working with are:

```python
df_tn.head(1)
```

```python
df_tn_mod = df_tn.set_index('movie')
```

```python
df_tn_mod.head(1)
```

```python
df_tmdb.info()
```

```python
df_tmdb.head(1)
```

```python
#making a modified copy of df_tmdb 
df_tmdb_mod = df_tmdb
```

```python
#copy 'orig_title' column into 'movie' column to line up with other dataframe
df_tmdb_mod['movie'] = df_tmdb_mod['original_title']

#setting index to movie in order to join
df_tmdb_mod = df_tmdb_mod.set_index('movie')

#dropping release date because it appears in the other database and was causing a conflict 
#(one of the release_date columns were note conveted to date/time***)
df_tmdb_mod = df_tmdb_mod.drop('release_date', axis=1)
```

```python
df_tmdb_mod.head(1)
```

### Left Join

```python
df_tmdb_tn_leftjoin = df_tmdb_mod.join(df_tn_mod, on='movie', how='left', lsuffix='tmbd', rsuffix='tn')
df_tmdb_tn_leftjoin
```

```python
len(set(df_tmdb_tn_leftjoin['title']))
```

### Inner Join

```python
df_tmdb_tn_innerjoin = df_tmdb_mod.join(df_tn_mod, on='movie', how='inner', lsuffix='tmbd', rsuffix='tn')
df_tmdb_tn_innerjoin
```

```python
#to get only unique rows for df, **i'm not sure why the inner join dindn't take care of this
df_tmdb_tn_innerjoin_unique = df_tmdb_tn_innerjoin.drop_duplicates()
df_tmdb_tn_innerjoin_unique
```

```python
df_tmdb_tn_innerjoin_unique.sort_values(by='id', ascending=False).tail(50)
```
